# Calculating the determinant of 2 3D vectors

```python
import numpy as np
import vector as vec
```

```python
# computing the signed volume
def volume(a,b,verbose = True):
    am = vec.mag(a)
    bm = vec.mag(b)

    dot = np.dot(a,b)
    cos = dot/(am*bm)
    sin = np.sqrt(1 - cos**2)

    if verbose:
        print(f"cos = {cos}, sin = {sin}")

    proj = dot/(am**2)*a
    error = b - proj
    em = vec.mag(error)
    volume_error = em*am
    volume_sine = am*bm*sin

    if verbose:
        print(f"volume_error = {volume_error}, volume_sine = {volume_sine}")

    return volume_error

```

```python
a = np.array([1,1,1])
b = np.array([1,1,0])
c = np.array([1,1,2])

am = vec.mag(a)
bm = vec.mag(b)
cm = vec.mag(c)
```
