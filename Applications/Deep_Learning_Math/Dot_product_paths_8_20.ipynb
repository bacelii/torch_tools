# Looking if inverse path is correct

```python
import numpy as np
M = np.array([
    [1,4,5],
    [2,10,8],
    [5,6,12]
])

ax,ay,az = M[:,0]
bx,by,bz = M[:,1]
cx,cy,cz = M[:,2]

byr = by - ay/ax*bx
bzr = bz - az/ax*bx
cyr = cy - ay/ax*cx
czr = cz - cyr*bzr/byr - cx*az/ax



U_non_norm = np.array([
    [ax,bx,cx],
    [0,byr,cyr],
    [0,0,czr],
])

L = np.array([
    [1,0,0],
    [ay/ax,1,0],
    [az/ax,bzr/byr,1]
])

U = np.array([
    [1,bx/byr,cx/czr],
    [0,1,cyr/czr],
    [0,0,1]
])

D = np.diag([ax,byr,czr])
result = (L@U@D)#.astype('int')
display(result,M)
np.array_equal(result,M)
result - M
```

```python
# showing how to manually get the inverses
```

```python
L_inv = np.array([
    [1,0,0],
    [-ay/ax,1,0],
    [-az/ax + (-bzr/byr)*(-ay/ax),-bzr/byr,1]
])

L_inv_2_step = np.array([
    [1,0,0],
    [0,1,0],
    [0,-bzr/byr,1]
])@np.array([
    [1,0,0],
    [-ay/ax,1,0],
    [-az/ax,0,1]
])

display(L@L_inv)
display(L@L_inv_2_step)

U_inv = np.array([
    [1, -bx/byr, -cx/czr + bx/byr*cyr/czr],
    [0,1,-cyr/czr],
    [0,0,1]
])

U_inv_2_step = np.array([
    [1,-bx/byr,0,],
     [0,1,0,],
    [0,0,1]
]) @ np.array([
    [1,0,-cx/czr],
    [0,1,-cyr/czr],
    [0,0,1]
])

display(U@U_inv)
display(U@U_inv_2_step)
```
