# Calculating the Determinant

```python
import numpy as np
import vector as vec
```

```python
vecs = np.array([
    [1,3,8],
    [2,10,3],
    [-1,2,4]
])
M = vecs.T
M
```

```python
def gram_schmidt_columns(X):
    Q, R = np.linalg.qr(X)
    return Q,R

Q,R = gram_schmidt_columns(M)
Q,R
```

```python
det_by_hand = np.prod(np.diag(R))
det_by_np = np.linalg.det(M)

det_by_hand,det_by_np
```

# Determine if Determinant can be understood by its projectinos

```python
import numpy as np

a = np.array([1,1,3])
b = np.array([1,2,4])
c = np.array([1,3,7])

M = np.vstack([a,b,c]).T
M
```

```python
x,y,z = list(range(3))

def proj_area(x,y):
    return (
        a[x]*b[y] - b[x]*a[y]
    + a[x]*c[y] - c[x]*a[y]
    + b[x]*c[y] - c[x]*b[y]
    )
proj_area(0,1),proj_area(0,2),proj_area(1,2)
```
