```python
"""
Purpose: to plot the determinant of two unit vectors as a function of the angle between them and
see if there is a specific relationship

"""
```

```python
import numpy as np
```

```python
v1 = np.array([1,0.5])
v2 = np.array([1,0])


v1 = v1/np.linalg.norm(v1)
v2 = v2/np.linalg.norm(v2)

def rot_matrix(theta):
    return np.array([
        [np.cos(theta), -1*np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])

def determinant(v1,v2):
    return v1[0]*v2[1] - v1[1]*v2[0]

n = 1000
thetas = np.linspace(0,2*np.pi,n)
```

```python
import matplotlib.pyplot as plt

v2 = v2.reshape(-1,1)
det = [determinant(v1,rot_matrix(t)@v2)
        for t in thetas]
```

```python
fig,ax = plt.subplots(1,1)

ax.plot(thetas,det,"r")
#ax.plot(thetas,np.sin(thetas),"b")

plt.show()
```
